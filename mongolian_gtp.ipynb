# **Self Attention Head**

self_attention.gif

```python
import torch.nn as nn
from torch.nn import functional as F
```

```python
class Head(nn.Module):
    """one head of self attention"""
    
    #n_embed embedding size, нэг head-ийн хувьд манай үгнүүд хэр хэмжээтэй векторт хөрвөхийг шийднэ
    #head_size embedded векторуудыг дахин өөр хэмжээтэй хөрвүүлэх
    #max_seq максимум үгийн цувааны хэмжээ 

    def __init__(self, n_embed, head_size, max_seq, dropout):
        super().__init__()
        self.key = nn.Linear(n_embed, head_size, bias=False)
        self.query = nn.Linear(n_embed, head_size, bias=False)
        self.value = nn.Linear(n_embed, head_size, bias=False)
        self.register_buffer('tril', torch.tril(torch.ones(max_seq, max_seq)))

        self.dropout = nn.Dropout(dropout)


    def forward(self, x):
        B,T,C = x.shape
        k = self.key(x) # (B, T, head_size) OR B,T,C    
        q = self.query(x) # (B, T, head_size) OR B,T,C
        
        #compute the attention weights
        weights = q @ k.transpose(-2,-1) * C**-0.5 # (B,T,C) @ (B,C,T) = (B,T,T)
        #mask out the future tokens
        weights = weights.masked_fill(self.tril[:T, :T] == 0, float('-inf')) # (B,T,T)
        
        #softmax to get the attention weights
        weights = weights.softmax(dim=-1) # (B,T,T)
        weights = self.dropout(weights)
        
        #compute the weighted sum of the values 
        v = self.value(x) # (B,T,C)
        out = weights @ v # (B,T,T) @ (B,T,C) = (B,T,C)
        
        return out
```

# **MultiHead Attention**

```python
class MultiHeadAttention(nn.Module):
    def __init__(self, num_heads, embed_size, max_seq, dropout=0):
        
        #num_heads нийт хэдэн self-attention head байхыг сонгоно
        #embed_size embedding size, манай үгнүүд хэр хэмжээтэй векторт хөрвөхийг шийднэ
        
        super().__init__()
        
        head_embed_size = embed_size // num_heads
        
        #multihead attention
        self.heads = nn.ModuleList([Head(embed_size, head_embed_size, 
                                         max_seq, dropout) for _ in range(num_heads)]) 
        
        self.ff = nn.Linear(head_embed_size*num_heads, embed_size)
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, inputs):
        
        # batch, ctx, embed_size
        inputs = torch.cat([head(inputs) for head in self.heads], 
                           dim=-1) # batch, ctx, num_head*head_size --> batch, ctx, embed_size
        
        inputs = self.ff(inputs) # batch, ctx, embed_size
        inputs = self.dropout(inputs)
        return inputs
```

# **Transformer Block**

```python
class Block(nn.Module):
    
    def __init__(self, num_heads, embed_size, max_seq, dropout=0):
        super().__init__()
        
        self.ln1 = nn.LayerNorm(embed_size)  #layer norm
        
        self.mha = MultiHeadAttention(num_heads, embed_size, max_seq, dropout) #selft attention
        
        self.ln2 = nn.LayerNorm(embed_size) #layer norm
        
        ## MLP layer
        self.ff = nn.Sequential(
            nn.Linear(embed_size, embed_size),
            nn.ReLU(),
            nn.Linear(embed_size, embed_size),
            nn.Dropout(dropout)
        )
    
    def forward(self, inputs):
        inputs = self.mha(self.ln1(inputs)) + inputs # batch, ctx, embed_size
        inputs = self.ff(self.ln2(inputs)) + inputs  # batch, ctx, embed_size
        return inputs
```

# **Language Model**

```python
class LanguageModel(nn.Module):
    def __init__(self):
        
        #VOCAB_SIZE давтагдаагүй үгийн тоо
        # 
        
        super().__init__()
        
        self.embedding = nn.Embedding(VOCAB_SIZE, embed_size)
        self.pos_embedding = nn.Embedding(max_seq, embed_size)
        self.blocks = nn.Sequential(*[Block(num_heads, embed_size, max_seq) for _ in range(num_layers)])
        self.ln = nn.LayerNorm(embed_size)
        self.ff = nn.Linear(embed_size, VOCAB_SIZE)
    
    def forward(self, inputs, targets=None):
        # batch, ctx, vocab_size
        tok_emb = self.embedding(inputs) # batch, ctx, embed_size
        pos_emb = self.pos_embedding(torch.arange(inputs.size(1), device=device))
        logits = tok_emb + pos_emb
        logits = self.blocks(logits) # batch, ctx, embed_size
        logits = self.ln(logits) # batch, ctx, embed_size
        logits = self.ff(logits) # batch, ctx, vocab_size

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            lv = logits.view((B*T, C))
            tv = targets.view((B*T,))
            loss = F.cross_entropy(lv, tv)
        
        return logits, loss

    def generate(self, tokens, n=1):
        for _ in range(n):
            logits,_ = self(tokens[:,-max_seq:])
            logits = logits[:,-1,:]
            logits = F.softmax(logits, dim=-1)
            next_token = torch.multinomial(logits, num_samples=1)
            tokens = torch.cat((tokens, next_token), dim=1)
        return tokens
```

# **Model Training**

```python
batch_size = 32 # how many independent sequences will be processed in parallel
eval_interval = 500 # how often to evaluate the model
n_heads = 6 # number of attention heads
n_layers = 6 # number of layers
dropout = 0.2 # dropout rate
VOCAB_SIZE = len(words)
embed_size = 384 
max_seq = 128
num_layers=4
num_heads=4
lr = 0.0001
steps=2000

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

torch.manual_seed(23)
```

```python
#train and test split
data = list_of_words
n = int(len(data) * 0.9)
train_data = data[:n]
val_data = data[n:]
```

```python
#title
model = LanguageModel()
print(sum(p.numel() for p in model.parameters())/1e6, 'M parameters')
model.to(device)
model.train()

optim = torch.optim.Adam(model.parameters(), lr=lr)

for step in range(steps):

  X, y = get_batch(train_data, encode, max_seq, batch_size)
  
  inter_perc = (step % eval_interval) / eval_interval
  avancement = '=' * int(30*inter_perc) + '>' + '_' * int(30*(1-inter_perc))
  print(f"\rStep {step:5} {avancement:33} {inter_perc*100:3.1f}%", end="")

  if step % eval_interval == 0:
      model.eval()
      with torch.no_grad():
          losses = torch.zeros(100)
          for k in range(100):
              X, y = get_batch(val_data, encode, max_seq, batch_size)
              logits, loss = model(X, y)
              losses[k] = loss.item()
          val_loss = losses.mean()
      print(f"\rStep {step:5}: val loss {val_loss:.5f}")
      
      if step % eval_interval == 0:
          encoded = torch.tensor(encode(['Миний', 'Улаанбаатар']), dtype=torch.int64, device=device).unsqueeze(0)
          print("\t", decode(model.generate(encoded, 32)[0].cpu().numpy()))
      
      model.train()

  _, loss = model(X, y)
  optim.zero_grad(set_to_none=True)
  loss.backward()
  optim.step()
```
